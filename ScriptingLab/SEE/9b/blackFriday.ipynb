```python
import pandas as pd
import numpy as np

black = pd.read_csv("BlackFriday.csv")
black.head()
```

```python
black.describe()
```

```python
black['City_Category'] = black['City_Category'].fillna("default")
black.head()
```

```python
black["City_Category"] = black["City_Category"].map({
    "A": "Metro Cities",
    "B": "Small Towns",
    "C": "Villages"
})

black.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax = sns.countplot(x="City_Category", hue="Gender", data=black)
ax.set(title="city vs gender", xlabel="City_Category", ylabel="Total")
plt.show()
```
