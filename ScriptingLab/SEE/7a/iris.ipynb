```python
import pandas as pd
import numpy as np

iris = pd.read_csv("iris.csv")
iris.head()
```

```python
iris.describe()
```

```python
copy = iris.drop("Class", axis= 1, inplace = False)
copy.head()
```

```python
mean = iris[["Class", " Petal_Width"]]
mean.groupby(["Class"], as_index = False).mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize = [12, 6])
ax = sns.countplot(x=" Sepal_Width", hue="Class", data=iris)
ax.set(title='Flowers of each specie',xlabel='Sepal Width',ylabel='No. of flowers')
plt.show()
```
